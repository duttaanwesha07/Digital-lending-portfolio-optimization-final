```python
import pandas as pd
pd.set_option("display.width", 200)
pd.set_option("display.max_columns", 30)

# Path to the data folder (canonical flattened path; WS E nesting removed during reconciliation)
DATA_DIR = "../WS E - Customer Segmentation"

# Load loan-level data
lf = pd.read_parquet(f"{DATA_DIR}/out/frames/loan_frame.parquet")
print(f"Loaded {len(lf):,} loans")
print(f"Columns: {len(lf.columns)} total")

# Load segment assignments from Step 3
seg = pd.read_csv(f"{DATA_DIR}/step3_segment_assignments.csv")
print(f"Loaded {len(seg):,} segment assignments")

# Merge so each loan has its segment name attached
lf = lf.merge(seg[["loan_id", "segment"]], on="loan_id")
print(f"\nAfter merge: {len(lf):,} loans with segment column")
print("\nSegment counts:")
print(lf["segment"].value_counts().sort_index())
```

```python
# Recompute total value_proxy by segment from the loan_frame
recon = (lf.groupby("segment")["value_proxy"].sum() / 1e7).round(2)
recon = recon.reset_index().rename(columns={"value_proxy": "recomputed_cr"})

# Load the step3 baseline metrics for comparison
step3 = pd.read_csv(f"{DATA_DIR}/step3_heuristic_baseline_metrics.csv")
compare = step3[["segment", "total_value_proxy_cr"]].merge(recon, on="segment")
compare["diff"] = (compare["recomputed_cr"] - compare["total_value_proxy_cr"]).round(2)
print(compare)
```

```python
# BNPL by acquisition channel — basis for Lever 1
bnpl = lf[lf["product_type"] == "BNPL"]
print(f"Total BNPL loans: {len(bnpl):,}")

channel_view = bnpl.groupby("acquisition_channel").agg(
    n_loans=("loan_id", "count"),
    mean_value_proxy=("value_proxy", "mean"),
    total_value_proxy_cr=("value_proxy", lambda x: round(x.sum() / 1e7, 2)),
    default_rate_pct=("default_flag", lambda x: round(x.mean() * 100, 2)),
).reset_index().sort_values("mean_value_proxy")

print("\nBNPL by acquisition channel (worst on top):")
print(channel_view)
```

```python
# LEVER 1 — Cease BNPL via Digital ads and DSA
# Net Δ at baseline = -(sum of value_proxy of removed loans)

lever1_targets = bnpl[bnpl["acquisition_channel"].isin(["Digital ads", "DSA"])]

n_affected = len(lever1_targets)
ticket_cr = lever1_targets["ticket_size"].sum() / 1e7
current_vp_cr = lever1_targets["value_proxy"].sum() / 1e7
net_delta_cr = -current_vp_cr  # negate because removing losses gains the book

print(f"Lever 1 — Cease BNPL via Digital ads + DSA")
print(f"  Loans removed:           {n_affected:,}")
print(f"  Origination given up:    ₹{ticket_cr:.2f} cr")
print(f"  Current value_proxy:     ₹{current_vp_cr:.2f} cr  (negative = book is currently losing this much)")
print(f"  Net Δ at baseline:       +₹{net_delta_cr:.2f} cr  (this is what the book gains)")
```

```python
# Lever 1 stress testing
# A-044: LGD = 40% for BNPL (locked data-generating value, config.py)
# A003: Macro stress band = +2pp / +3pp

LGD_BNPL = 0.40

def avoided_stress_loss(target_loans, stress_pp, lgd):
    """Additional loss the ceased loans would have taken under stress — we avoid this."""
    return (target_loans["ticket_size"] * lgd * stress_pp / 100).sum() / 1e7

baseline_net = -lever1_targets["value_proxy"].sum() / 1e7
extra_2pp = avoided_stress_loss(lever1_targets, 2, LGD_BNPL)
extra_3pp = avoided_stress_loss(lever1_targets, 3, LGD_BNPL)

print(f"Lever 1 — Cease BNPL via Digital ads + DSA")
print(f"  Net Δ at 0pp (baseline):    +₹{baseline_net:.2f} cr")
print(f"  Additional gain at +2pp:    +₹{extra_2pp:.2f} cr")
print(f"  Additional gain at +3pp:    +₹{extra_3pp:.2f} cr")
print()
print(f"  TOTAL Net Δ at +2pp stress:  +₹{baseline_net + extra_2pp:.2f} cr")
print(f"  TOTAL Net Δ at +3pp stress:  +₹{baseline_net + extra_3pp:.2f} cr")
```

```python
# Reusable lever computation for any CEASE/TIGHTEN lever
def compute_cease_lever(target_loans, lever_name, lgd):
    """
    `target_loans` = the slice of the loan_frame we'd remove from the book.
    Returns baseline + stressed net Δ in ₹ cr.
    """
    baseline = -target_loans["value_proxy"].sum() / 1e7
    extra_2 = (target_loans["ticket_size"] * lgd * 0.02).sum() / 1e7
    extra_3 = (target_loans["ticket_size"] * lgd * 0.03).sum() / 1e7

    print(f"{lever_name}")
    print(f"  Loans affected:           {len(target_loans):,}")
    print(f"  Origination given up:     ₹{target_loans['ticket_size'].sum()/1e7:.2f} cr")
    print(f"  Net Δ at baseline (0pp):  +₹{baseline:.2f} cr")
    print(f"  Net Δ at +2pp stress:     +₹{baseline + extra_2:.2f} cr")
    print(f"  Net Δ at +3pp stress:     +₹{baseline + extra_3:.2f} cr")
    return {
        "name": lever_name,
        "n": len(target_loans),
        "ticket_cr": target_loans["ticket_size"].sum() / 1e7,
        "baseline_cr": baseline,
        "stress_2pp_cr": baseline + extra_2,
        "stress_3pp_cr": baseline + extra_3,
    }

# Verify by re-running Lever 1 through the function
lever1_result = compute_cease_lever(
    lever1_targets, "Lever 1 — Cease BNPL Digital + DSA", LGD_BNPL
)
```

```python
# Filter to Personal D-E
ppl_de = lf[(lf["product_type"] == "Personal") & 
            (lf["origination_risk_grade"].isin(["D", "E"]))]

print(f"Personal D-E total loans:         {len(ppl_de):,}")
print(f"Personal D-E mean value_proxy:    ₹{ppl_de['value_proxy'].mean():,.0f}")
print(f"Personal D-E total value_proxy:   ₹{ppl_de['value_proxy'].sum()/1e7:.2f} cr")
print(f"Personal D-E % loss-making:       {(ppl_de['value_proxy'] < 0).mean()*100:.1f}%")

# Band loans by cashflow_consistency to see where they go loss-making
bands = pd.cut(
    ppl_de["cashflow_consistency_mean"],
    bins=[0, 0.3, 0.4, 0.5, 0.6, 0.7, 1.0],
    labels=["<0.3", "0.3-0.4", "0.4-0.5", "0.5-0.6", "0.6-0.7", "0.7+"]
)

band_view = ppl_de.groupby(bands, observed=True).agg(
    n_loans=("loan_id", "count"),
    mean_value_proxy=("value_proxy", "mean"),
    total_vp_cr=("value_proxy", lambda x: round(x.sum()/1e7, 2)),
    default_rate_pct=("default_flag", lambda x: round(x.mean()*100, 2)),
).reset_index()

print("\nPersonal D-E by cashflow_consistency band (lowest = worst cashflow):")
print(band_view)
```

```python
# Lever 2 — Tighten Personal D-E with cashflow_consistency < 0.6
LGD_PERSONAL = 0.65  # Personal — unsecured retail (locked data-generating value, config.py A-044)

lever2_targets = ppl_de[ppl_de["cashflow_consistency_mean"] < 0.6]

lever2_result = compute_cease_lever(
    lever2_targets,
    "Lever 2 — Tighten Personal D-E (cashflow_consistency < 0.6)",
    LGD_PERSONAL
)
```

```python
# Filter to SME C-D-E
sme_cde = lf[(lf["product_type"] == "SME") & 
             (lf["origination_risk_grade"].isin(["C", "D", "E"]))]

print(f"SME C-D-E total loans:          {len(sme_cde):,}")
print(f"SME C-D-E mean value_proxy:     ₹{sme_cde['value_proxy'].mean():,.0f}")
print(f"SME C-D-E total value_proxy:    ₹{sme_cde['value_proxy'].sum()/1e7:.2f} cr")
print(f"SME C-D-E % loss-making:        {(sme_cde['value_proxy'] < 0).mean()*100:.1f}%")

# Band by cashflow_consistency to find the cutoff
bands = pd.cut(
    sme_cde["cashflow_consistency_mean"],
    bins=[0, 0.3, 0.4, 0.5, 0.6, 0.7, 1.0],
    labels=["<0.3", "0.3-0.4", "0.4-0.5", "0.5-0.6", "0.6-0.7", "0.7+"]
)

band_view = sme_cde.groupby(bands, observed=True).agg(
    n_loans=("loan_id", "count"),
    mean_value_proxy=("value_proxy", "mean"),
    total_vp_cr=("value_proxy", lambda x: round(x.sum()/1e7, 2)),
    default_rate_pct=("default_flag", lambda x: round(x.mean()*100, 2)),
).reset_index()

print("\nSME C-D-E by cashflow_consistency band:")
print(band_view)
```

```python
# Lever 5 — Contain SME C-D-E with cashflow_consistency < 0.60 (SUPERSEDED by < 0.65 below)
LGD_SME = 0.75  # SME (locked data-generating value, config.py A-044)

lever5_targets = sme_cde[sme_cde["cashflow_consistency_mean"] < 0.60]

lever5_result = compute_cease_lever(
    lever5_targets,
    "Lever 5 — Contain SME C-D-E (cashflow_consistency < 0.60)",
    LGD_SME
)
```

```python
# Lever 5 — Contain SME C-D-E with cashflow_consistency < 0.65
### post discussion edit: relaxed cutoff to <0.65 (empirical optimum per WS H §4)
LGD_SME = 0.75  # SME (locked data-generating value, config.py A-044)

lever5_targets = sme_cde[sme_cde["cashflow_consistency_mean"] < 0.65]

lever5_result = compute_cease_lever(
    lever5_targets,
    "Lever 5 — Contain SME C-D-E (cashflow_consistency < 0.65)",
    LGD_SME
)
```

```python
# Lever 4 — Grow SME A-B by 10%
# Different math from Cease levers: ADD new loans rather than remove.
# Assumption: new loans look like the existing SME A-B average.

GROWTH_PCT = 10  # Conservative — can flex later in sensitivity analysis

sme_ab = lf[(lf["product_type"] == "SME") & 
            (lf["origination_risk_grade"].isin(["A", "B"]))]

n_current = len(sme_ab)
n_new = int(round(n_current * GROWTH_PCT / 100))
mean_vp = sme_ab["value_proxy"].mean()
mean_ticket = sme_ab["ticket_size"].mean()

# Baseline: each new loan brings the segment's mean value_proxy
baseline_cr = n_new * mean_vp / 1e7

# Stress: new loans ALSO take additional default losses under stress
extra_2pp = n_new * mean_ticket * LGD_SME * 0.02 / 1e7
extra_3pp = n_new * mean_ticket * LGD_SME * 0.03 / 1e7

stress_2pp_cr = baseline_cr - extra_2pp  # subtract: more losses HURT the gain
stress_3pp_cr = baseline_cr - extra_3pp

lever4_result = {
    "name": f"Lever 4 — Grow SME A-B by {GROWTH_PCT}%",
    "n": n_new,
    "ticket_cr": n_new * mean_ticket / 1e7,
    "baseline_cr": baseline_cr,
    "stress_2pp_cr": stress_2pp_cr,
    "stress_3pp_cr": stress_3pp_cr,
}

print(f"Lever 4 — Grow SME A-B by {GROWTH_PCT}%")
print(f"  Current SME A-B loans:          {n_current:,}")
print(f"  New loans added:                {n_new:,}")
print(f"  Mean value_proxy per loan:      ₹{mean_vp:,.0f}")
print(f"  Mean ticket per loan:           ₹{mean_ticket:,.0f}")
print(f"  Origination added:              ₹{n_new * mean_ticket / 1e7:.2f} cr")
print(f"  Net Δ at baseline (0pp):        +₹{baseline_cr:.2f} cr")
print(f"  Net Δ at +2pp stress:           +₹{stress_2pp_cr:.2f} cr")
print(f"  Net Δ at +3pp stress:           +₹{stress_3pp_cr:.2f} cr")
```

```python
# Build the consolidated scenarios table from all four lever results

# Tag each lever with the sign of its volume change (Cease=-, Grow=+)
lever_signs = {
    "Lever 1 — Cease BNPL Digital + DSA": -1,
    "Lever 2 — Tighten Personal D-E (cashflow_consistency < 0.6)": -1,
    "Lever 4 — Grow SME A-B by 10%": +1,
    "Lever 5 — Contain SME C-D-E (cashflow_consistency < 0.65)": -1,
}

all_levers = [lever1_result, lever2_result, lever4_result, lever5_result]
macro_levels = [
    ("0pp (baseline)", "baseline_cr"),
    ("+2pp", "stress_2pp_cr"),
    ("+3pp", "stress_3pp_cr"),
]

rows = []
# Per-lever rows
for lever in all_levers:
    sign = lever_signs[lever["name"]]
    for macro, net_key in macro_levels:
        rows.append({
            "Scenario": lever["name"],
            "Macro": macro,
            "Volume Δ (₹ cr)": round(sign * lever["ticket_cr"], 2),
            "Net Δ (₹ cr)": round(lever[net_key], 2),
        })

# Combined portfolio rows
for macro, net_key in macro_levels:
    total_vol = sum(lever_signs[l["name"]] * l["ticket_cr"] for l in all_levers)
    total_net = sum(l[net_key] for l in all_levers)
    rows.append({
        "Scenario": "COMBINED (all 4 levers)",
        "Macro": macro,
        "Volume Δ (₹ cr)": round(total_vol, 2),
        "Net Δ (₹ cr)": round(total_net, 2),
    })

scenarios = pd.DataFrame(rows)
print(scenarios.to_string(index=False))

# Save for the workbook
scenarios.to_csv("scenarios_table.csv", index=False)
print("\nSaved: scenarios_table.csv (in WS G folder)")
```
